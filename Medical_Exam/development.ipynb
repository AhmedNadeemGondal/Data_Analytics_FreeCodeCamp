```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('medical_examination.csv')
df_drop_bp = df[df['ap_lo'] >= df['ap_hi']]
df_drop_bp
```

```python
(df_drop_bp['height'] >= df_drop_bp['height'].quantile(0.025)) & (df_drop_bp['height'] <= df_drop_bp['height'].quantile(0.975))
# df_drop_bp['height'].quantile(0.025)
# df_drop_bp['height'].quantile(0.975)
df_drop_bp_h = df_drop_bp[(df_drop_bp['height'] >= df_drop_bp['height'].quantile(0.025)) & (df_drop_bp['height'] <= df_drop_bp['height'].quantile(0.975))]
df_drop_bp_h
```

```python
df_drop_bp_hw = df_drop_bp[(df_drop_bp['weight'] >= df_drop_bp['weight'].quantile(0.025)) & (df_drop_bp['weight'] <= df_drop_bp['weight'].quantile(0.975))]
df_drop_bp_hw
```

```python
df_drop_bp_hw['overweight'] = df_drop_bp_hw['weight'] / ((df_drop_bp_hw['height']/100) ** 2)
df_drop_bp_hw['overweight'] = (df_drop_bp_hw['overweight'] <= 25).astype(int)
df_drop_bp_hw
# df_drop_bp_hw['overweight'] <= 25
```

```python

df_drop_bp_hw.loc[df_drop_bp_hw['cholesterol'] == 1, 'cholesterol'] = 0
df_drop_bp_hw.loc[df_drop_bp_hw['cholesterol'] > 1, 'cholesterol'] = 1

df_drop_bp_hw['cholesterol'].value_counts()
```

```python
df_drop_bp_hw.loc[df_drop_bp_hw['gluc'] == 1, 'gluc'] = 0
df_drop_bp_hw.loc[df_drop_bp_hw['gluc'] > 1, 'gluc'] = 1

```

```python
df_drop_bp_hw.head()
```

```python
melted_df = pd.melt(df_drop_bp_hw, id_vars=['cardio'], value_vars=['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke'], var_name='variable', value_name='total')
melted_df.head()
# ['cardio','id','age','sex','height','weight','ap_hi','ap_lo']
```

```python
# Create the chart
sns.catplot(data=melted_df, kind='count', x='variable', hue='total', col='cardio', col_wrap=2)

# Show the chart
plt.show()
```

```python
df_heat = df_drop_bp_hw.corr()
masked_corr_matrix = df_heat.mask(np.triu(np.ones_like(df_heat, dtype=bool)))
masked_corr_matrix
```

```python
# plt.figure(figsize=(10, 8))
sns.heatmap(data=masked_corr_matrix, annot=True, fmt=".1f", cmap="coolwarm",cbar_kws={"shrink": 0.8})
plt.show()
```
